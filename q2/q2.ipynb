```python
import numpy as np
import matplotlib.pyplot as plt

def calcAve(distribution):
    return np.average(distribution)

def calcVariance(distribution):
    return np.var(distribution)

def createHistogram(distribution, title, bins=30):
    plt.hist(distribution, bins=bins, edgecolor='k')
    plt.title(title)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
```

## (1) 一様分布
### コード

```python
def uniformDistribution(low, high, size=1000):
    return np.random.uniform(low, high, size)

zeroToOne = uniformDistribution(0, 1)
print("0から1の一様分布の平均値: ", calcAve(zeroToOne))
print("0から1の一様分布の平均値(厳密解):", (0 + 1) / 2)
print("0から1の一様分布の分散: ", calcVariance(zeroToOne))
print("0から1の一様分布の分散(厳密解):", (1 - 0) ** 2 / 12)
createHistogram(zeroToOne, 'Uniform Distribution from 0 to 1')
towToFive = uniformDistribution(2, 5)
print("2から5の一様分布の平均値: ", calcAve(towToFive))
print("2から5の一様分布の平均値(厳密解):", (2 + 5) / 2)
print("2から5の一様分布の分散: ", calcVariance(towToFive))
print("2から5の一様分布の分散(厳密解):", (5 - 2) ** 2 / 12)
createHistogram(towToFive, 'Uniform Distribution from 2 to 5')
```

### 考察
0~1までの一様分布と2~5までの一様分布を求めたが、どちらも一様分布になっていることがわかる。  
また、得られた分散と平均は、理論値と比較すると0.1~0.3程度の誤差があり、ヒストグラムからも比較的平らな形状になっているが、誤差があることがわかる。

## (2) 指数分布
### コード

```python
def exponentialDistribution(lam, size=1000):
    return np.random.exponential(1 / lam, size)

lamOne = exponentialDistribution(1)
print("λ=1の指数分布の平均値: ", calcAve(lamOne))
print("λ=1の指数分布の平均値(理論値):", 1 / 1)
print("λ=1の指数分布の分散: ", calcVariance(lamOne))
print("λ=1の指数分布の分散(理論値):", (1 / 1) ** 2)
createHistogram(lamOne, 'Exponential Distribution λ=1')

lamEleven = exponentialDistribution(11)
print("λ=11の指数分布の平均値: ", calcAve(lamEleven))
print("λ=11の指数分布の平均値(理論値):", 1 / 11)
print("λ=11の指数分布の分散: ", calcVariance(lamEleven))
print("λ=11の指数分布の分散(理論値):", (1 / 11) ** 2)
createHistogram(lamEleven, 'Exponential Distribution λ=11')
```

### 考察
今回はλ=1の指数分布を求めたが、ヒストグラムより指数分布の特徴である、右に裾が長い形状になっていることがわかる。
また、λ=1やλ=11のときに得られた分散と平均は、理論値と比較すると誤差があり、ヒストグラムにもその影響が出ていることがわかる。

## (4) 正規分布
### コード

```python
def normalDistribution(mu, sigma, size=1000):
    x = np.random.uniform(mu, sigma, size)
    y = np.random.uniform(mu, sigma, size)
    return np.sqrt(-2 * np.log(x)) * np.cos(2 * np.pi * y)

muZeroSigmaOne = normalDistribution(0, 1)
print("平均0, 分散1の正規分布の平均値: ", calcAve(muZeroSigmaOne))
print("平均0, 分散1の正規分布の平均値(理論値):", 0)
print("平均0, 分散1の正規分布の分散: ", calcVariance(muZeroSigmaOne))
print("平均0, 分散1の正規分布の分散(理論値):", 1)
createHistogram(muZeroSigmaOne, 'Normal Distribution μ=0, σ=1')

muZeroSigmaFour = normalDistribution(0, 4)
print("平均0, 分散5の正規分布の平均値: ", calcAve(muZeroSigmaFour))
print("平均0, 分散5の正規分布の平均値(理論値):", 0)
print("平均0, 分散5の正規分布の分散: ", calcVariance(muZeroSigmaFour))
print("平均0, 分散5の正規分布の分散(理論値):", 4)
createHistogram(muZeroSigmaFour, 'Normal Distribution μ=0, σ=5')
```

### 考察
今回は正規分布を求めたが、ヒストグラムより正規分布の特徴である、左右対称な形状になっていることがわかる。  
また、得られた分散と平均は、理論値と比較すると誤差があり、ヒストグラムにもその影響が出ていることがわかる。

## (5) ポアソン分布
### コード

```python
def poissonDistribution(lam, size=100000):
    return np.random.poisson(lam, size)

lamFive = poissonDistribution(5)
print("λ=5のポアソン分布の平均値: ", calcAve(lamFive))
print("λ=5のポアソン分布の平均値(理論値):", 5)
print("λ=5のポアソン分布の分散: ", calcVariance(lamFive))
print("λ=5のポアソン分布の分散(理論値):", 5)
createHistogram(lamFive, 'Poisson Distribution λ=5', bins=15)

lamTen = poissonDistribution(10)
print("λ=10のポアソン分布の平均値: ", calcAve(lamTen))
print("λ=10のポアソン分布の平均値(理論値):", 10)
print("λ=10のポアソン分布の分散: ", calcVariance(lamTen))
print("λ=10のポアソン分布の分散(理論値):", 10)
createHistogram(lamTen, "Poisson Distribution λ=10", bins=15)
```

### 考察
ポアソン分布なので、ヒストグラムは左右非対称な形状になっている。
また、λ=5やλ=10のときに得られた分散と平均は、理論値と比較すると誤差があり、ヒストグラムにもその影響が出ていることがわかる。

## (6) ベータ分布
### コード

```python
def betaDistribution(a, b, size=1000):
    return np.random.beta(a, b, size)

aOneBOne = betaDistribution(1, 1)
print("a=1, b=1のベータ分布の平均値: ", calcAve(aOneBOne))
print("a=1, b=1のベータ分布の平均値(理論値):", 1 / (1 + 1))
print("a=1, b=1のベータ分布の分散: ", calcVariance(aOneBOne))
print("a=1, b=1のベータ分布の分散(理論値):", (1 * 1) / ( (1 + 1) ** 2 *(1 + 1 + 1)))
createHistogram(aOneBOne, 'Beta Distribution a=1, b=1')

aTwoBThree = betaDistribution(2, 3)
print("a=2, b=3のベータ分布の平均値: ", calcAve(aTwoBThree))
print("a=2, b=3のベータ分布の平均値(理論値):", 2 / (2 + 3))
print("a=2, b=3のベータ分布の分散: ", calcVariance(aTwoBThree))
print("a=2, b=3のベータ分布の分散(理論値):", (2 * 3) / ( (2 + 3) ** 2 *(2 + 3 + 1)))
createHistogram(aTwoBThree, 'Beta Distribution a=2, b=3')
```

### 考察
若干の誤差はあるが、理論値に近い値と形状になっていることがわかる。

```python
def chisquareDistribution(n,size=1000):
    return np.random.chisquare(n, size)

nFour = chisquareDistribution(4)
print("自由度4のカイ二乗分布の平均値: ", calcAve(nFour))
print("自由度4のカイ二乗分布の平均値(理論値):", 4)
print("自由度4のカイ二乗分布の分散: ", calcVariance(nFour))
print("自由度4のカイ二乗分布の分散(理論値):", 2 * 4)
createHistogram(nFour, 'Chisquare Distribution n=4')

nTen = chisquareDistribution(10)
print("自由度10のカイ二乗分布の平均値: ", calcAve(nTen))
print("自由度10のカイ二乗分布の平均値(理論値):", 10)
print("自由度10のカイ二乗分布の分散: ", calcVariance(nTen))
print("自由度10のカイ二乗分布の分散(理論値):", 2 * 10)
createHistogram(nTen, 'Chisquare Distribution n=10')
```
